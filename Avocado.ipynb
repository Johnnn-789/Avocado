```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set_theme(style="whitegrid")
df = pd.read_csv('avocado.csv').dropna()
df
```

```python
sns.countplot(data=df, x='year')
plt.title('Total number of weekly paperwork reports!')
plt.show()
```

```python
counts = df['type'].value_counts()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title('Percentage of types of Avocado report!')
plt.show()
```

```python
sns.histplot(data=df, x='AveragePrice', bins=20)
plt.title('Average Price Distribution')
plt.show()
```

```python
sns.kdeplot(data=df, x='AveragePrice', fill=True)
plt.title('Average Price Density Curve')
plt.show()
```

```python
sns.histplot(data=df, x='AveragePrice', bins=20, kde=True)
plt.title('Average Price Distribution with KDE')
plt.show()
```

```python
sns.lineplot(x='year', y='Total Volume', data=df)
plt.show()
```
